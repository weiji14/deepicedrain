```python
def test_plot_datetime():
    """Test various datetime input data"""
    import datetime
    fig = pygmt.Figure()
    fig.basemap(projection="X15c/5c", region="2010-01-01/2020-01-01/0/10", frame=True)

    # numpy.datetime64 types
    x = np.array(
        ["2010-06-01", "2011-06-01T12", "2012-01-01T12:34:56"], dtype="datetime64"
    )
    y = [1.0, 2.0, 3.0]
    fig.plot(x, y, style="c0.2c", pen="1p")

    # pandas.DatetimeIndex
    x = pd.date_range("2013", freq="YS", periods=3)
    y = [4, 5, 6]
    fig.plot(x, y, style="t0.2c", pen="1p")

    # xarray.DataArray
    x = xr.DataArray(data=pd.date_range(start="2015-03", freq="QS", periods=3))
    y = [7.5, 6, 4.5]
    fig.plot(x, y, style="s0.2c", pen="1p")

    # raw datetime strings
    x = ["2016-02-01", "2017-03-04T00:00"]
    y = [7, 8]
    fig.plot(x, y, style="a0.2c", pen="1p")

    # the Python built-in datetime and date
    x = [datetime.date(2018, 1, 1), datetime.datetime(2019, 1, 1)]
    y = [8.5, 9.5]
    fig.plot(x, y, style="i0.2c", pen="1p")
    return fig

f = test_plot_datetime()
f.savefig("figures/test_plot_datetime.png")
f.show()
```

```python
@pygmt.helpers.use_alias(
    R="region",
    J="projection",
    Jz="zscale",
    JZ="zsize",
    A="straight_lines",
    B="frame",
    C="cmap",
    D="position",
    E="error_bars",
    G="color",
    L="close",
    S="style",
    W="pen",
    i="columns",
    l="label",
    p="perspective",
)
@pygmt.helpers.kwargs_to_strings(R="sequence", i="sequence_comma", p="sequence")
def plot3d(x, y, z, **kwargs):
    """Thin wrapper around pygmt.plot3d"""
    extra_arrays = []
    with pygmt.clib.Session() as lib:
        file_context = lib.virtualfile_from_vectors(
            np.atleast_1d(x), np.atleast_1d(y), np.atleast_1d(z), *extra_arrays
        )
        with file_context as fname:
            arg_str = " ".join([fname, pygmt.helpers.build_arg_string(kwargs)])
            lib.call_module("plot3d", arg_str)
```

```python
idx = 26991341
# idx = np.argmax(ds_dhdt.dhdt_slope)
ds_many: xr.Dataset = ds_dhdt.isel(ref_pt=slice(int(ds_dhdt.ref_pt[idx])-10, int(ds_dhdt.ref_pt[idx])+10))
# ds_many: xr.Dataset = ds_.isel(ref_pt=slice(int(ds_.ref_pt[idx])-100000, int(ds_.ref_pt[idx])+100000))
df: pd.DataFrame = ds_many.to_dataframe()
df = df.dropna()
```

```python
# Convert delta_time to utc_time, and get dates
df["utc_time"] = deepicedrain.deltatime_to_utctime(df.delta_time)
dates = np.datetime_as_string(df.utc_time, unit="ns")
```

```python
# Plot 2D along track
fig = pygmt.Figure()
#region = [dates[3], dates[4], 0, 2, np.nanmin(z)-0.1, np.nanmax(z)+0.1]
#region = [dates[0], dates[-1], np.nanmin(z)-0.1, np.nanmax(z)+0.2]

region = [df.x_atc.min(), df.x_atc.max(), df.h_corr.min()-0.2, df.h_corr.max()+0.2]

fig.basemap(
    projection="X20c/10c",
    region=region,
    frame=['WSne+t"Change in height along an ICESat-2 track"', 'xaf+l"Along track x (m)"', 'yaf+l"Height (m)"'],
    #p="225/30",
    #Jz="0.0001"
)
pygmt.makecpt(cmap="no_green", series=(3, 8, 1))
for cycle, color in zip((3, 5, 6, 7), ("#984ea3", "#4daf4a", "#377eb8", "#e41a1c")):
    df_ = df.query(expr="cycle_number == @cycle").copy()
    if len(df_) > 0:
        fig.plot(x=df_.x_atc, y=df_.h_corr, style="c0.1c", pen="1p") # Plot data points
        fig.plot(x=df_.x_atc, y=df_.h_corr, pen=f"2p,{color}", label=f'"Cycle {cycle}"') # Plot line
    #fig.plot(x=x, y=y, pen="1p,darkgray,-")  # Regression line
fig.legend(S=1)
fig.savefig("figures/dh_along_track_example.png")
fig.show()
```

```python
import numpy as np
import pandas as pd
import datashader as ds
import datashader.transfer_functions as tf
from collections import OrderedDict as odict

num=100000
np.random.seed(1)

dists = {cat: pd.DataFrame(odict([('x',np.random.normal(x,s,num)), 
                                  ('y',np.random.normal(y,s,num)), 
                                  ('val',val), 
                                  ('cat',cat)]))      
         for x,  y,  s,  val, cat in 
         [(  2,  2, 0.03, 10, "d1"), 
          (  2, -2, 0.10, 20, "d2"), 
          ( -2, -2, 0.50, 30, "d3"), 
          ( -2,  2, 1.00, 40, "d4"), 
          (  0,  0, 3.00, 50, "d5")] }

df = pd.concat(dists,ignore_index=True)
df["cat"]=df["cat"].astype("category")
```

```python
hv.notebook_extension('bokeh')
```

```python
points = hv.Points(df.sample(10000))

hv.operation.datashader.datashade(points)
```

```python
hv.output(backend="bokeh")
hd.datashade(points)
```

```python
import holoviews.operation.datashader as hd
from holoviews.streams import RangeXY

pts = hd.datashade(points, width=400, height=400)

quadmesh =  hv.QuadMesh(hd.aggregate(points, width=10, height=10, dynamic=False)) \
            .opts(tools=['hover'], alpha=0, hover_alpha=0.2)

dynamic = hv.util.Dynamic(hd.aggregate(points, width=10, height=10, streams=[RangeXY]), 
                          operation=hv.QuadMesh) \
          .opts(tools=['hover'], alpha=0, hover_alpha=0.2)

(pts * quadmesh).relabel("Fixed hover") + (pts * dynamic).relabel("Dynamic hover")
```
